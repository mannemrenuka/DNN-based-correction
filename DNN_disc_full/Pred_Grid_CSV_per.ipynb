```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from keras.layers import Dropout
from keras.optimizers import RMSprop
from sklearn.cross_validation import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import math
from keras.models import load_model
import os
import pandas
from scipy.io import loadmat

sub=['F1','F2','M1','M2']
# sub = ['M2']
fold=['Fold1','Fold2','Fold3','Fold4']
# fold = ['Fold1']
seg = 'Lower'
# a = list(range(99,100))
b = [99.75,100]
per = b
print('Processing!!!')
for p in per:
    for s in sub:
#         np_val = numpy.loadtxt(seg+'/'+s+'/'+"np_fold_"+str(p)+".csv", delimiter=",")
        mat = loadmat('../DNN_Data_dist_DNN_Full/'+seg+'/'+s+'/'+"pxl_np_"+str(p)+".mat")
        np_fold = mat['pxl_np']
        for f in fold:
            print(s+' and '+ f + ' and ' + str(p))
            model = load_model(seg+'/'+s+'/'+f+'/'+'Model_GridWise_dev_'+str(p)+'.hdf5')
#             dataset = pandas.read_csv(seg+'/'+s+'/'+f+'/'+"Grid_Wise_Test_"+str(p)+".csv", delimiter=",",header = None)            
#             dataset.fillna(dataset.mean(), inplace=True)
#             dataset = dataset.as_matrix(columns = None)
            dataset = pandas.read_csv(seg+'/'+s+'/'+f+'/'+"Dev_Intensity_Data_"+str(p)+".csv", delimiter=",",header = None)
            dataset.fillna(dataset.mean(), inplace=True)
            dataset = dataset.as_matrix(columns = None)
            ind = fold.index(f)
            np = np_fold[0,ind][1][0][0]
            print(np)
            possi = []
            for i in range(0,np+1):
                possi.append(numpy.round(model.predict(dataset[:,i*(np+1):(i+1)*(np+1)]),decimals=3))
#             X1=dataset[:,0:np+1]
#             X2=dataset[:,np+1:2*(np+1)]
#             X3=dataset[:,2*(np+1):3*(np+1)]
#             Y1=numpy.round(model.predict(X1),decimals=3)
#             Y2=numpy.round(model.predict(X2),decimals=3)
#             Y3=numpy.round(model.predict(X3),decimals=3)
#             a = numpy.asarray([ Y1, Y2, Y3 ])
            possi = numpy.squeeze(numpy.asarray(possi))
            numpy.savetxt(seg+'/'+s+'/'+f+'/'+"Pred_Data_Grid_"+str(p)+".csv", possi, delimiter=",")
            possi = []
            dataset = []
            mat = []
print('Done!!!')
```
