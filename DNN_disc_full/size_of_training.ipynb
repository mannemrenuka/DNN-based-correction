```python
import numpy
import pandas
from scipy.io import loadmat
sub=['F1','F2','M1','M2']
fold=['Fold1','Fold2','Fold3','Fold4']
# sub = ['F1']
# fold = ['Fold1']
seg='Upper'
# per = [95,96,97,98,99,99.25,99.5,99.75,99.85,99.9,99.95,100];
a = list(range(60,100))
b = [99.25,99.5,99.75,100]
per = a + b
# per = a
# per = [80]
l = numpy.zeros(len(per)*len(sub)*len(fold))
MSE=[]
c1 = 0
for p in per:
    for s in sub:
        mat = loadmat('../DNN_Data_dist_DNN_Full/'+seg+'/'+s+'/'+"pxl_np_"+str(p)+".mat")
        np_fold = mat['pxl_np']
        for f in fold: 
            ind = fold.index(f)
            np = np_fold[0,ind][1][0][0]
            print('perc'+str(p)+'Subject'+s+'and'+f)
            dataset = loadmat(seg+'/'+s+'/'+f+'/'+"Intensity_Set_"+str(p)+".mat")
            a = dataset['intensity_profile_frame']
            stacked_arr = [];
            for b in a[0]:
                for c in b:
                    stacked_arr = stacked_arr  + [i for i in c if i!=[]]
            stacked_arr = numpy.array([stacked_arr])
            stacked_arr = numpy.squeeze(stacked_arr)
            stacked_arr.shape
            l[c1] = len(stacked_arr)
            c1 = c1+1
```
